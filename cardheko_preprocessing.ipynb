### km

```python
km=[]
for i in data["km"]:
    km_str=i.replace(",",'').strip()
    km.append(int(km_str))

a=pd.DataFrame(km,columns=["km"])

data["km"]=a["km"]
```

### "price" data cleaning

```python
amounts = []
for i in data["price"]:
    amount_str = i.replace('₹ ', '').replace(' Lakh', '').replace(' Crore', '').replace(',', '').strip()
    if 'Crore' in i:
        amounts.append(float(amount_str) * 100)  # Convert crores to lakhs
    elif 'Lakh' in i:
        amounts.append(float(amount_str))  
    else:
        amounts.append(float(amount_str)/100000)

price = pd.DataFrame(amounts, columns=['price in lakhs'])


data["price (lakh)"] = price["price in lakhs"]
```

### Max Power

```python
data["Max Power"].unique()
```

```python
mode=data["Max Power"].mode()[0]
data["Max Power"].fillna(mode,inplace=True)
```

```python
import re
a=[]
count=-1
for i in data["Max Power"]:
    count+=1
    i_str= re.split(r'bhp|@|at|-| PS |PS| |HP|B|Ps|ps', i)[0]
    i_str = re.sub(r"hp", "", i_str)
    try:
        a.append(float(i_str))
    except:
        a.append(float(i_str.split('(')[0]))
```

```python
Max_Power = pd.DataFrame(a, columns=['Max_Power'])
data["Max Power"] = Max_Power["Max_Power"]
```

### Torque

```python
mode=data["Torque"].mode()[0]
data["Torque"].fillna(mode,inplace=True)
```

```python
torque=[]
count=-1
for i in data["Torque"]:
    count+=1
    try:
        torque_str=re.split(r'Nm|nm|NM|/', i)[0]
        #print(torque_str)
        torque.append(float(torque_str))
    except:
        #print(data.loc[count,"Torque"])
        if "(" in i:
            torque_str=re.split(r' ', i)[0]
            #print(torque_str)
            torque.append(float(torque_str))
        else:
            torque_str=re.split(r' ', i)[0]
            #print(f"{float(torque_str)*9.80665:.2f}")
            torque.append(float(f"{float(torque_str)*9.80665:.2f}"))

# 1 kgm = 9.80665 Nm
```

```python
Torque=pd.DataFrame(torque, columns=['Torque'])
data["Torque"]=Torque["Torque"]
```
