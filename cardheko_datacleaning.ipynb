# new_car_feature

```python
empty_list=[]
for i in df['new_car_feature']:
    temp = ast.literal_eval(i)
    temp = temp["data"]
    emptylist={}
    emptylist["total car features"]= len(temp)
    for j in temp:
        emptylist[j["heading"]]= "yes"
    empty_list.append(pd.DataFrame([emptylist]))
```

```python
empty_list
```

```python
new_car_feature = pd.concat(empty_list, ignore_index=True)
```

```python
new_car_feature
```

```python
new_car_feature.fillna("no",inplace=True)
```

```python
new_car_feature
```

# new_car_specs

```python
specs_list=[]

for i in df["new_car_specs"]:
    temp = ast.literal_eval(i)
    emptylist={}
   # print(temp)
    for j in temp["top"]:
        #print(j)
        emptylist[j['key']]= j['value']
        
    for j in temp["data"]:
        #print(j)
        for k in j["list"]:
            #print(k)
            emptylist[k['key']]= k['value']
    #print(emptylist)

    
    specs_list.append(pd.DataFrame([emptylist]))

    
```

```python
specs_list
```

```python
new_car_specs = pd.concat(specs_list, ignore_index=True)
```

```python
new_car_specs
```

# new_car_overview

```python
features_list = []

# Process top features
for i in df["new_car_overview"]:
    temp = ast.literal_eval(i)
    temp=temp["top"]
    emptylist={}
    
    for j in temp:
        emptylist[j['key']]= j['value']
    print(emptylist)
    features_list.append(pd.DataFrame([emptylist]))

# Create a DataFrame from the features list
#features_df = pd.DataFrame(features_list)

# Display the resulting DataFrame
features_list
```

```python
features_list
```

```python
new_car_overview = pd.concat(features_list, ignore_index=True)
```

```python
new_car_overview
```
